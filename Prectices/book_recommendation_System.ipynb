### Collaborative-based Recommendation System

**Approach:**

Our collaborative-based recommendation system aims to leverage user ratings on various books to provide personalized recommendations. The system utilizes two primary datasets: rating data and book data.

**Datasets:**

1. **Rating Data:**
   - Contains user ratings for different books.
   - Columns: `User-ID`, `ISBN`, `Book-Rating`

2. **User Data:**
   - Includes information about users such as location and age.
   - Columns: `User-ID`, `Location`, `Age`

3. **Book Data:**
   - Provides details about books including title, author, publication year, and publisher.
   - Columns: `ISBN`, `Book-Title`, `Book-Author`, `Year-Of-Publication`, `Publisher`, `Image-URL-S`, `Image-URL-M`, `Image-URL-L`

**Methodology:**

1. **Data Integration:**
   - We merge the rating data with user data and book data using common identifiers (`User-ID` and `ISBN`). This consolidation facilitates a comprehensive analysis.

2. **User-Item Matrix Construction:**
   - A pivotal step involves constructing a user-item matrix where rows correspond to users, columns represent books, and each cell captures the rating provided by a user for a specific book.
   - As not all users would have rated all books, the matrix tends to be sparse.

3. **Similarity Calculation:**
   - Utilizing measures like cosine similarity or Pearson correlation coefficient, we compute similarities between users or items.
   - By analyzing ratings given by users to different books, we establish similarity between users based on their preferences and tendencies.

4. **Recommendation Generation:**
   - Given a target user, we identify similar users based on calculated similarity scores.
   - Recommendations are then generated by suggesting books highly rated by similar users but not yet rated by the target user.
   - Techniques such as neighborhood-based collaborative filtering or matrix factorization can be employed for recommendation generation.

**Evaluation:**

   - To assess the effectiveness of our recommendation system, we employ metrics such as precision, recall, or mean average precision.
   - The dataset is typically split into training and testing sets to evaluate the system's performance accurately.

```python
users_rating = new_data.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
users_high_rating = users_rating[users_rating].index
users_high_rating
```

```python
filter_rating = new_data[new_data['User-ID'].isin(users_high_rating)]
filter_rating
```

```python
top_50_users_rated_book = filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = top_50_users_rated_book[top_50_users_rated_book].index
```

```python
famous_books
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
final_rating.drop_duplicates()
```

```python
pt =final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

### 📊 Similarity Score Analysis:

Here's a snippet of the similarity score array, which measures the similarity between each book and all other books in the dataset. Let's break it down:

- Each element in the array corresponds to the similarity score between the book at index 0 and every other book in the dataset.
- A similarity score of 1.0 indicates a perfect match (the book compared with itself).
- Lower similarity scores signify less similarity between books, while higher scores indicate greater similarity.
- The similarity scores provide valuable insights into which books are most closely related to each other in terms of user ratings.

By analyzing these similarity scores, we can identify books with similar characteristics and recommend them to users based on their preferences. This enables us to create a robust recommendation system tailored to the interests of our audience. 📚🔍

## Now let's Convert in fuction

```python
def recommender(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(enumerate(similarity_score[index]), key=lambda x: x[1], reverse=True)[1:7]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommender('The Rainmaker')
```

```python
recommender('A Heartbreaking Work of Staggering Genius')
```
